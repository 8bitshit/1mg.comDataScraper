```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as soup
```

```python
header = {'Origin': 'https://www.1mg.com',
          'Referer': 'https://www.1mg.com/categories/exclusive/immunity-boosters/vitamin-c-734',
          'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36'}
```

```python
url = 'https://www.1mg.com/categories/exclusive/immunity-boosters/vitamin-c-734'
```

```python
html = requests.get(url = url, headers = header)
html.status_code  #checking if request was successful, status code must evaluate to 200, otherwise request has failed
```

```python
bsobj = soup(html.content, 'html.parser')
bsobj
```

```python
prod_name_output = bsobj.findAll('div', {'class': 'style__pro-title___2QwJy'})
pack_size_output = bsobj.findAll('div', {'class': 'style__pack-size___2JQG7'})
prod_price_output = bsobj.findAll('div', {'class': 'style__price-tag___cOxYc'})
```

```python
product_name = []
for name in prod_name_output:
    product_name.append(name.text.strip())

product_name
```

```python
package_size = []
for name in pack_size_output:
    package_size.append(name.text.strip())

package_size
```

```python
product_price = []
for price in prod_price_output:
    product_price.append(format(float(price.text.replace('â‚¹', '').strip()), '.2f'))

product_price
```

```python
df_dict = {'pname': product_name, 
           'psize': package_size,
           'pprice': product_price}
```

```python
df = pd.DataFrame.from_dict(df_dict)
df = df.rename(columns={'pname': 'Product name', 'psize': 'Package size', 'pprice': 'Price'})
df
```
